````python
# input()

name = input("Enter your name: ")
age = input("Enter your age: ")
height = input("Enter your height: ")
favourite_number = input("Enter your Favourite Number: ")

print(f"Welcome {name}")
print(f"Your age is {age}")
print(f"Your height is {height}")
print(f"Your Favourite Number is {favourite_number}")
````

# Personal Data Collector (Python)

## 📌 Project Description
This is a beginner-level Python project that collects personal information from the user using the `input()` function and displays it using the `print()` function.

The project demonstrates the basic concepts of Python such as:
- Variables
- Data Types
- User Input
- Formatted Output (f-strings)

---

## 🚀 Features

- Takes user's name
- Takes user's age
